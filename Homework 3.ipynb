```python
import pandas as pd
import nltk
from nltk import pos_tag, word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer

training_dataset = pd.read_csv('train.txt', delimiter="\n", header=None)

# print(training_dataset)

def convert_to_present_tense(word):
    return WordNetLemmatizer().lemmatize(word,'v')

ctr = 0
for line in training_dataset[0]:
    line_arr = line.split()
    statement = " ".join(line_arr[1:])
    print(statement)
    text = word_tokenize(statement)
    result = nltk.pos_tag(text)
    
    print(result)
    
    for i in range(len(result)):
        if("VB" in result[i][1]):
            result[i] = (convert_to_present_tense(result[i][0]), "VB")
    
    print(result)
    
    ctr += 1
    if (ctr == 100):
        break
```
